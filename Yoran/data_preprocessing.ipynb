```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./responses_raw.csv')
df = df.replace(np.nan, '', regex=True)
```

```python
df['kindergarten_open_to_vital'] = df['open_to_vital'].str.contains('Kindergarten')
df['highschool_open_to_vital'] = df['open_to_vital'].str.contains('Highschool')
df['university_open_to_vital'] = df['open_to_vital'].str.contains('University')
```

```python
categorical_array_columns = ['prohibited_all', 'open_to_vital', 'closed_to_everyone']
```

```python
for category in categorical_array_columns:
    unique = np.unique([item for sublist in df[category].map(lambda x: (x.split(', '))).tolist() for item in sublist]).tolist()
    for key in unique:
        df_key = key.replace(' ', '_').lower() + "_" + category
        df[df_key] = df[category].str.contains(key)
        df[df_key] = df[df_key].astype(int)
df = df.drop(categorical_array_columns, axis=1)
```

```python
categorical_columns = ['sufficient_protection', 'mentality_category']
df = pd.concat([df, pd.get_dummies(df['sufficient_protection'], prefix='protection')], 1)
df = df.drop(categorical_columns, axis=1)
```
