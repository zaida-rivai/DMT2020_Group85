```python
weather_raw_20200120 = weather_raw.loc[(weather_raw['date'] >= '2020-01-20') & (weather_raw['date'] <= '2020-05-14')]
weather_raw_20200120
```

```python
weather_20200120_drop = weather_raw_20200120.drop(['code','province','precipitation','max_wind_speed','most_wind_direction'],axis=1)
weather_20200120_drop_g = weather_20200120_drop.groupby('date').mean()
weather_20200120_drop_g
```

```python
weather_20200120_grouped = weather_20200120_drop_g.reset_index()
weather_20200120_grouped
```
