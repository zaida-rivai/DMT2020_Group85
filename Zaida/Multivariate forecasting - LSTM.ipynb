```python
dfConfirmed = pd.read_csv("time_series_covid19_confirmed_global.csv")
dfDeath = pd.read_csv("time_series_covid19_deaths_global.csv")
dfRecovered = pd.read_csv("time_series_covid19_recovered_global.csv")

cases = [dfConfirmed,dfDeath,dfRecovered]
```

```python
df['Timestamp']= pd.to_datetime(df['Timestamp']) 
df['Timestamp'].apply(lambda x: x.strftime('%m/%d'))
```

##### preprocess JH dataset

```python
countryNamesQ = list(df.country)
```

```python
def preprocessJh(df,countryNamesQ):
    df = df[df['Province/State'].isnull()]
    df = df.drop(['Province/State','Lat', 'Long'],axis=1)
    df = df[df['Country/Region'].isin(countryNamesQ)]
    return df

for i in range(len(cases)): 
    cases[i] = preprocessJh(cases[i],countryNamesQ)
```
